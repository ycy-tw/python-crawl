### 臺灣上市公司股票價格2(不含公司債)
### Taiwan listed stock's price(exclude corp bonds)

```python
from pandas.core.frame import DataFrame
import requests
import pandas as pd

def get_stock_price(date:str) -> DataFrame:

    url = 'https://www.twse.com.tw/exchangeReport'
    url += '/MI_INDEX?response=json&date=20210603&type=ALL'

    res = requests.get(url)
    source = res.json()

    df = pd.DataFrame(source['data9'], columns=source['fields9'])
    df = df[df['證券代號'].str.len() == 4]
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
get_stock_price('20210601')
```
