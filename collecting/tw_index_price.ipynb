### 臺灣證券交易所各類型指數日內報價(5秒)
### Taiwan's indices intraday quote information.

```python
from pandas.core.frame import DataFrame
import pandas as pd


def taiex_index(date: str) -> DataFrame:

    # set target url
    url = 'https://www.twse.com.tw/exchangeReport'
    url += f'/MI_5MINS_INDEX?response=csv&date={date}'
    table = pd.read_table(url, encoding='big5')
    data = list(table[table.columns[0]].str.split(',"'))

    """
    input:
    data[-5:]

    output:
    [['說明:'],
    ['民國100年1月16日以前，資料為每一分鐘方式提供。'],
    ['民國103年2月23日以前，資料為每15秒方式提供。'],
    ['民國103年2月24日以後，資料為每10秒方式提供。'],
    ['民國103年12月29日以後，資料為每5秒方式提供。']]
    """
    df = pd.DataFrame(data)[0:-5] # remove additional information
    df = df.apply(
        lambda x: x
        .str.replace('"', '')
        .str.replace('=', '')
        .str.replace(',', '')
    )
    df.columns = df.loc[0]
    df = df[1:]
    df.set_index('時間', inplace=True, drop=True)
    df = df.astype(float)

    return df
```
